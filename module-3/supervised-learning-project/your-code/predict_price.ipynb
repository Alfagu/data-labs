```python
import pandas as pd
import numpy as np
```

```python
cars = pd.read_csv('./cars_test.csv')
cars.head()
```

```python
cars['cylinders'] = cars['cylinders'].map({'other': 0, 
                                           '3 cylinders': 3, 
                                           '4 cylinders': 4, 
                                           '5 cylinders': 5, 
                                           '6 cylinders': 6, 
                                           '8 cylinder': 8, 
                                           '10 cylinders': 10, 
                                           '12 cylinders': 12})
cars['fuel'] = cars['fuel'].map({'other': 0, 
                                 'gas': 1, 
                                 'diesel': 2, 
                                 'hybrid': 3,
                                 'electric': 4})
cars['condition'] = cars['condition'].map({'salvage': 0, 
                                           'fair': 1, 
                                           'good': 2, 
                                           'like new': 3, 
                                           'excellent': 4, 
                                           'new': 5})
cars = cars.dropna()

cars.head()
```
