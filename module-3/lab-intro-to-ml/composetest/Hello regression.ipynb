```python
import pandas as pd
```

```python
from sklearn.datasets import load_boston

boston = load_boston()
```

```python
b = pd.DataFrame(boston.data, columns=boston.feature_names)
b['PRICE'] = boston.target
b.head()
```

```python
b.describe()
```

```python
b.dtypes
```

```python
X = b.drop(['PRICE','CRIM', 'INDUS', 'B'], axis=1)
Y = b['PRICE']
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.33, random_state=5)
```

```python
from sklearn.linear_model import LinearRegression

regressor = LinearRegression()

regressor.fit(X_train, Y_train)
regressor.score(X_test, Y_test)
```

```python
from sklearn.ensemble import RandomForestRegressor

regressor = RandomForestRegressor(random_state=5)

regressor.fit(X_train, Y_train)
regressor.score(X_test, Y_test)
```

```python
from sklearn.neighbors import KNeighborsRegressor

regressor = KNeighborsRegressor()

regressor.fit(X_train, Y_train)
regressor.score(X_test, Y_test)
```

```python
import pandas as pd
from sklearn.datasets import load_diabetes

diabetes = load_diabetes()
```

```python
b = pd.DataFrame(diabetes.data, columns=diabetes.feature_names)
b['age'] = diabetes.target
b.head()
```

```python
b.describe()
```

```python
X = b.drop(['bmi'], axis=1)
Y = b['age']
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.33, random_state=5)
```

```python
from sklearn.linear_model import LinearRegression

regressor = LinearRegression()

regressor.fit(X_train, Y_train)
regressor.score(X_test, Y_test)
```

```python
from sklearn.ensemble import RandomForestRegressor

regressor = RandomForestRegressor(random_state=5)

regressor.fit(X_train, Y_train)
regressor.score(X_test, Y_test)
```

```python
from sklearn.neighbors import KNeighborsRegressor

regressor = KNeighborsRegressor()

regressor.fit(X_train, Y_train)
regressor.score(X_test, Y_test)
```
