## Now we visualize how much high was the Sales Prices through the years in the data base. And we can see head to head the Sale Prices from different types of Sale.

### The conclusions we can take from this quick analysis: are that the most mayority of the tickets sales are from the WD type. Followed by the New Type and the third in this case will be COD type. 

```python
data.groupby(['YrSold', 'SaleType']).sum().unstack().plot(y='SalePrice', kind='bar', stacked=True, figsize=(10, 10))
```
